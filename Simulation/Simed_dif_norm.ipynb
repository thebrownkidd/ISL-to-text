```python
def toarray(a):
    a = str(a)
    a = a[1:len(a)-2]
    val = []
    x=0
    while x<len(a):
        if(len(val) == 21):
            break
        temp = ""
        # print(a[x:])
        for y in range(x,len(a)):
            if a[y] != ",":
                temp += a[y]
            else:
                x = y+1
                break
        if len(temp) != 0:
            val.append(float(temp))
        x+=1
    return val
```

```python
def dist(x1,y1,x2,y2):
    dist = ((x1-x2)**2 + (y1-y2)**2)**0.5
    return dist
```

```python
Data = pd.read_csv('C:\Projects\ISL-to-text\Data\Centralized_Simulator.csv')
base = Data.iloc[0,1:]
final = []
for i in range(1,63):
    diff_x = [0]
    diff_y = [0]
    # print("Entry: ",i)
    for j in range(1,21):
        # print("\tElement: ",j)
        if j not in (5,9,13,17):
            d = dist(toarray(base[0])[j],toarray(base[1])[j],toarray(base[0])[j-1],toarray(base[1])[j-1])
        else:
            d = dist(toarray(base[0])[j],toarray(base[1])[j],toarray(base[0])[0],toarray(base[1])[0])
        diff_x.append((toarray(base[0])[j]- toarray(Data.iloc[i,1])[j])/d)
        diff_y.append((toarray(base[1])[j]- toarray(Data.iloc[i,2])[j])/d)
    final.append([diff_x,diff_y])
print(final)
```
