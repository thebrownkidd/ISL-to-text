```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
def toarray(a):
    a = a[1:len(a)-2]
    val = []
    x=0
    while x<len(a):
        if(len(val) == 21):
            break
        temp = ""
        # print(a[x:])
        for y in range(x,len(a)):
            if a[y] != ",":
                temp += a[y]
            else:
                x = y+1
                break
        if len(temp) != 0:
            val.append(float(temp))
        x+=1
    return val
```

```python
def separr(s):
    a = ''
    arr = []
    for i in range(1,len(s)-2):
        if s[i] != ']':
            if s[i] == '[':
                i += 1
            else:
                a += (s[i])
        else:
            arr.append(toarray(a))
            a = ''
    arr.append(toarray(a))
    return arr[0],arr[1]
```

```python
def one_hot(i):
    ret = [0]*9
    if i <= 5:
        ret[i] = 1
    else:
        ret[i-1] = 1
    return ret
```

```python
def soft_one_hot(Y):
    ret = []
    for x in Y:
        temp = [0]*9
        # x = pd.DataFrame(x)
        # print(temp.index(max(x)))
        temp[np.argmax(x)] = 1
        ret.append(temp)
    return ret
```

```python
import os
X = []
Y = []
dir = 'C:/Projects/ISL-to-text/0-9/'
for x in os.listdir(dir):
    g = []
    f = open(dir+x+"/asl"+x+".jpgXY.txt",'r')
    a = f.readline()
    m,n = separr(a)
    Y.append(one_hot(int(x)))
    for j in range(len(m)):
        print(type(m[j]))
        g.append(m[j])
        g.append(n[j])
    X.append(g)
X_val = pd.DataFrame(X)
Y_val = pd.DataFrame(Y)
# print(type(X_val.iloc[?0,1]))
# print("\n\n\n\n")
# print(len(Y_val))
```

```python

train_data = pd.read_csv('C:\Projects\ISL-to-text\Data\simed_train_data.csv')
X = []
Y = []
for i in range(len(train_data)):
    x = toarray(train_data.iloc[i,1])
    y = toarray(train_data.iloc[i,2])
    temp = []
    # print(type(x[1]))
    for j in range(len(x)):
        temp.append(x[j])
        temp.append(y[j])
    X.append(temp)
    Y.append(one_hot(train_data.iloc[i,3]))

X_train = pd.DataFrame(X)
Y_train = pd.DataFrame(Y)
print(X_train)
print(Y_train)
# print(X)
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units = 42, activation = 'tanh',input_shape = [42]),
                            #  tf.keras.layers.Dense(units = 300,activation = 'tanh'),
                            #  tf.keras.layers.Dense(units = 240,activation = 'tanh'),
                            #  tf.keras.layers.Dense(units = 105, activation = 'tanh'),
                            #  tf.keras.layers.Dense(units= 120, activation = 'tanh'),
                            #  tf.keras.layers.Dense(units = 80, activation= 'tanh'),
                            #  tf.keras.layers.Dense(units = 50, activation='tanh'),
                             tf.keras.layers.Dense(units= 27, activation= 'tanh'),
                            #  tf.keras.layers.Dense(units = 40, activation='tanh'),
                            #  tf.keras.layers.Dense(units = 9,activation = 'tanh'),
                             tf.keras.layers.Dense(units = 9,activation= 'softmax')])
model.compile(optimizer='adam',loss='mae')
```

```python
losses = model.fit(X_train,Y_train,validation_data= (X_val,Y_val),batch_size = 558,epochs = 558)
```
